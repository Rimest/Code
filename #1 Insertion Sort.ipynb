```python
from random import randint
import matplotlib.pyplot as plt
import time
import numpy as np
```

```python
def Insertion_Sort(a):
    for i in range(1, len(a)):
        for j in range(i):
            if a[i] < a[j]:
                a[i], a[j] = a[j], a[i]
    return a
```

```python
array = [5, 1, 4, 2, 1, 7, 0]
```

```python
Insertion_Sort(array)
```

```python
N = 500
t = []
t1 = time.time()
for n in range(1, N + 1):
    #array = np.random.randint(-10, 10, size = n)
    #array = list(reversed(range(n)))
    start_time = time.time()
    Insertion_Sort(list(reversed(range(n))))
    time_ = time.time() - start_time
    t.append(time_)

print("--- %s seconds ---" % (time.time() - t1))
```

```python
plt.plot(range(1, N + 1), t)
```

```python
#Parabola func for approximation
def func(x, a, b, c):
    return a + b*x + c*x*x
sigma = np.array([1] * N)
x0 = np.array([0] * 3)
```

```python
#Approximation
import scipy.optimize as optimization
[coeff, mat] = optimization.curve_fit(func, np.array(range(1, N + 1)), np.array(t), x0, sigma)
```

```python
pol = func(range(1, 2 * N + 1), coeff[0], coeff[1], coeff[2])
```

```python
plt.plot(range(1, N + 1), t)
plt.plot(range(1, 2 * N + 1), pol)
```
